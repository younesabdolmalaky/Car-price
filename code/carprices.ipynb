```python
import numpy as np
import pandas as pd
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import r2_score
from sklearn.model_selection import train_test_split
from sklearn import metrics
import xgboost as xgb
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('../datasets/Car details v3.csv')
df
```

```python
df.isnull().sum()
```

## drop missing values

```python
df = df.dropna(axis=0)
```

## Correct mileage
The format of mileage is km/kg and kmpl, which we convert both to kmpl format for pre-processing.

```python
Correct_Mileage= []
for i in df.mileage:
    if str(i).endswith('km/kg'):
        i = i[:-6]
        i = float(i)*1.40
        Correct_Mileage.append(float(i))
    elif str(i).endswith('kmpl'):
        i = i[:-6]
        Correct_Mileage.append(float(i))
```

```python
df.loc[:,'mileage'] = Correct_Mileage
```

## Feature extraction
In the car model part, the first word is the car brand, and the car brand can be given as a feature to the regression algorithm

```python
df['Brand'] = df['name'].str.split(' ').str[0]
```

### Deleting extra characters at the end of the attribute value (like km/kg)


```python
df['engine'] = df['engine'].str.split(' ').str[0]
df['max_power'] = df['max_power'].str.split(' ').str[0]
df.drop(['name','torque'], inplace=True , axis=1)
```

```python
df.info()
```

### Convert string data types to numbers
In some features, because the measurement unit of the feature is at the end of the measured number, their data type is stored as a string, so we must convert them into numbers. And the next item is the categorical features that must be converted into numbers. To solve this problem, use OHE

```python
df['engine'] = df['engine'].astype(float)
df['max_power'] = df['max_power'].astype(float)
df.info()
```

```python
df = pd.get_dummies(df, columns = ['fuel', 'seller_type' , 'transmission' , 'owner' , 'Brand'])
df
```

```python
df.info()
```

### Determine the Target
We consider the selling price as a target and the rest of the features as independent features

```python
X = df.loc[:, df.columns != 'selling_price']
y = df['selling_price']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y)
```

```python
xg_reg = xgb.XGBRegressor(colsample_bytree=0.8, learning_rate=0.1, max_depth=6, n_estimators=1000)
```

```python
xg_reg.fit(X_train, y_train)
```

```python
y_pred = xg_reg.predict(X_test)
```

```python
metrics.mean_absolute_error(y_test, y_pred)
```

```python
metrics.r2_score(y_test, y_pred)
```
