```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('../datasets/Car details v3.csv')
df
```

```python
df.isnull().sum()
```

```python
df = df.dropna(axis=0)
```

```python
df.isnull().sum()
```

```python
Correct_Mileage= []
for i in df.mileage:
    if str(i).endswith('km/kg'):
        i = i[:-6]
        i = float(i)*1.40
        Correct_Mileage.append(float(i))
    elif str(i).endswith('kmpl'):
        i = i[:-6]
        Correct_Mileage.append(float(i))
df['mileage'] = Correct_Mileage
```

```python
df['Brand'] = df['name'].str.split(' ').str[0]
df['engine'] = df['engine'].str.split(' ').str[0]
df['max_power'] = df['max_power'].str.split(' ').str[0]
df.drop(['name','torque'], inplace=True , axis=1)
```

```python
df.info()
```

```python
df['engine'] = df['engine'].astype(float)
df['max_power'] = df['max_power'].astype(float)
df.info()
```

```python
df = pd.get_dummies(df, columns = ['fuel', 'seller_type' , 'transmission' , 'owner' , 'Brand'])
df
```

```python
df.info()
```

```python
X = df.loc[:, df.columns != 'selling_price']
y = df['selling_price']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)
```

```python
from sklearn.neighbors import KNeighborsRegressor
from sklearn.linear_model import LinearRegression
from sklearn.svm import SVR
Linear = LinearRegression()
svr_rbf = SVR(kernel="rbf", C=100, gamma=0.1, epsilon=0.1)
svr_lin = SVR(kernel="linear", C=100, gamma="auto")
svr_poly = SVR(kernel="poly", C=100, gamma="auto", degree=3, epsilon=0.1, coef0=1)
```

```python
Linear.fit(X_train, y_train)
svr_rbf.fit(X_train, y_train)
svr_lin.fit(X_train, y_train)
svr_poly.fit(X_train, y_train)
```

```python
y_pred_Linear = Linear.predict(X_test)
y_pred_svr_rbf = svr_rbf.predict(X_test)
y_pred_svr_lin = svr_lin.predict(X_test)
y_pred_svr_poly = svr_poly.predict(X_test)
```

```python
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import r2_score
```

```python
print('Linear' ,mean_absolute_error(y_test, y_pred_Linear) )
print('svr_rbf' ,mean_absolute_error(y_test, y_pred_svr_rbf) )
print('svr_lin' ,mean_absolute_error(y_test, y_pred_svr_lin) )
print('svr_poly' ,mean_absolute_error(y_test, y_pred_svr_poly) )
```

```python
print('Linear' ,r2_score(y_test, y_pred_Linear) )
print('svr_rbf' ,r2_score(y_test, y_pred_svr_rbf) )
print('svr_lin' ,r2_score(y_test, y_pred_svr_lin) )
print('svr_poly' ,r2_score(y_test, y_pred_svr_poly) )
```
